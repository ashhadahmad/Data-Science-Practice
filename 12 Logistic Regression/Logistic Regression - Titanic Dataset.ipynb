```python
import numpy as np
import pandas as pd
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
train_data = pd.read_csv('titanic_x_y_train.csv')
train_data.head(10)
```

```python
# Lets us fill the N.A values in Age column based on the Passenger class
def fillAge(cols):
    age = cols[0]
    cl  = cols[1]
    if pd.isnull(age):
#         return train_data[train_data["Pclass"] == cl]['Age'].mean()
        return train_data['Age'].mean()
    return age
```

```python
train_data['Age'] = train_data[['Age','Pclass']].apply(fillAge,axis=1)
```

```python
train_data.head(10)
```

```python
#Dropping Cabin Column

train_data.drop('Cabin',axis=1,inplace=True)
```

```python
train_data.head(10)
```

```python
train_data.describe()
```

```python
# Adding Gender as numerical value, 1 -> Female, 2 -> Male

train_data['gender'] = train_data.apply(lambda row: int(row['Sex'] == 'male') + 1, axis=1)
```

```python
train_data.head(10)
```

```python
train_data.isna().sum()
```

```python
#Since we still have one "NA", we can delete it
```

```python
train_data.dropna(inplace=True)
```

```python
train_data.isna().sum()
```

```python
train_data['Embarked'].value_counts()
```

```python
# embarked - Port of Embarkation (C = Cherbourg; Q = Queenstown; S = Southampton)

def getEmbarked(e):
    if e == 'S':
        return 1
    if e == 'C':
        return 2
    if e == 'Q':
        return 3

train_data['Emb'] = train_data['Embarked'].apply(getEmbarked)
```

```python
train_data.head(20)
```

Now we can remove Name, Sex, and Embarked because Name won't help the model, and we have already handled the other two

We also have to remove "Survived" column

```python
train_y = train_data["Survived"].values
train_y
```

```python
train_data.drop(['Survived', 'Name', 'Sex', 'Ticket','Embarked'], axis = 1, inplace = True)
train_data
```

```python
train_data
```

```python
algo = LogisticRegression(solver = 'liblinear')
```

```python
algo.fit(train_data, train_y)
```

```python
algo.score(train_data, train_y)
```

```python
allPages = ['http://books.toscrape.com/catalogue/page-1.html',
            'http://books.toscrape.com/catalogue/page-2.html']

column_names = ['Title', 'Link', 'Price', 'Quantity in Stock']

base_url = 'http://books.toscrape.com/'

from bs4 import BeautifulSoup
import pandas as pd
import requests
import re

books = []
for page in allPages:
    res = requests.get(page)
    soup = BeautifulSoup(res.text, 'html.parser')
    for book in soup.find_all(class_ = 'product_pod'):
        b = []
        title = book.h3.a['title']
        link = base_url + 'catalogue/' + book.h3.a['href']
        book_res = requests.get(link)
        book_soup = BeautifulSoup(book_res.text, 'html.parser')
        price = book_soup.find(class_ = 'price_color').string
        stock = list(book_soup.find(class_ = 'instock availability').strings)[-1].strip()
        price = float(re.search('[\d.]+', price).group())
        stock = int(re.search('\d+', stock).group())
        b.append(title)
        b.append(link)
        b.append(price)
        b.append(stock)
        books.append(b)
        print(*b)
df = pd.DataFrame(books, columns=['Title', 'Link', 'Price', 'Quantity in Stock'])
```

### Create two arrays, x and y. x should hold 20 even points starting from 0 and y should be an exponentiation array where y[i] = 2^x[i]. Plot a line graph, line should be of color blue and it should be a dashed line (like -----). Find the value of x from where there is slight increase in value of y as shown in your plot. Plot and print the value of x. Find the value of x from where there is slight increase in value of y as shown in your plot.

```python
import matplotlib.pyplot as plt
import numpy as np
x = np.arange(0, 40, 2, dtype = np.uint64)
y = 2 ** x
plt.plot(x,y,"b--")
plt.grid()
print(30)
```

### Plot the two graphs on same plot.The graph between the two given integer array X and Y and the graph between two integer array X and Y where X should hold 20 even points starting from 0 and y should be equal to X where y[i] = x[i].Plot should have x label and y label and the legend to differentiate between the two graphs.

```python
import matplotlib.pyplot as plt
import numpy as np
x1 =[0,1,2,3,4,5,6,7,9,10]
y1=[1,2,5,10,17,26,37,50,82,101]
x2 = np.arange(0,40,2)
y2 = x2
plt.plot(x1,y1, color = "red", label = "y = x + 1")
plt.plot(x2,y2, color = "blue", label = "y = x")
plt.grid()
plt.legend()
plt.xlabel("x axis")
plt.ylabel("y axis")
plt.title("Some cool graph")
plt.show()
```
